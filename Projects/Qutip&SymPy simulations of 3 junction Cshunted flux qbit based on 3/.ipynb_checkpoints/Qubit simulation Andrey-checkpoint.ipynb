```python
from qutip import *
from math import *
from scipy import constants as const
import numpy as np
from matplotlib import pyplot as plt
```

# Declaration of operators

```python
n = 3 # number of states in the Hilbert space
N = tensor(num(n), num(n)) # Number operator N1 * N2
N1 = tensor(num(n), qeye(n)) # Number operator in the 1st subspace
N2 = tensor(qeye(n), num(n)) # Number operator in the 2nd subspace
```

# Declaration of parameters

```python
Csh = 50e-15 # F
Ec = 17.6e9 # Hz
Ej = 84.18604651162792e9 # Hz
Cj = 2 * const.e ** 2 / Ec / const.h
```

# Potential energy

```python
x = np.linspace(-const.pi, const.pi, 100)
y = np.linspace(0, 5, 100)
xx, yy = np.meshgrid(x, y)
U = -Ej/2 * (np.cos(xx)*2 + yy * np.cos(2*xx + const.pi / 3))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, U/1e9)
ax.set_xlabel("$ \phi_1 = -\phi_2 $")
ax.set_ylabel(r"$\alpha$")
ax.set_zlabel("U, GHz")
plt.tight_layout()
plt.savefig("potential_energy.png")
plt.show()
```

# Custom functions

```python
# Square root of an operator
def qsqrt(obj):
    return Qobj(np.sqrt(obj.data))
```

```python
# Inverse of an operator
def qinv(obj):
    return Qobj(np.linalg.inv(obj.full()))
```

```python
# Extended tunneling with a coefficient
def x_tunneling(N, fi_ext):
    up = Qobj(np.diag(np.ones(N-1), -1))
    down = Qobj(np.diag(np.ones(N-1), 1))
    tun = np.exp(1j * fi_ext) * tensor(down, up)
    return tun + tun.dag()
```

# Constructing the Hamiltonian

```python
def getH(alpha, fi_ext):
    C = np.array([[Csh + (1 + alpha) * Cj, - Csh - alpha * Cj], [- Csh - alpha * Cj, Csh + (1 + alpha) * Cj]])
    Cinv = np.linalg.inv(C)
    Hc = 2 * const.e ** 2 * (Cinv[0, 0] * N1 * N1 + 2 * Cinv[0, 1] * N + Cinv[1, 1] * N2 * N2) / const.h
    cos_fi01 = tensor(tunneling(n, 1), qeye(n))
    cos_fi20 = tensor(qeye(n), tunneling(n, 1))
    cos_fi12 = x_tunneling(n, fi_ext)
    Hj = - Ej / 2 * (cos_fi01 + alpha * cos_fi12 + cos_fi20)
    return Hc + Hj
```

# Eigenenergies as a function of $\phi_{ext}$

```python
def calc_and_plot_Eigenenergies(alpha):
    fi_ext = np.linspace(4/5*const.pi, 6/5*const.pi, 101)

    E0 = np.empty_like(fi_ext)
    E1 = np.empty_like(fi_ext)
    E2 = np.empty_like(fi_ext)
    E3 = np.empty_like(fi_ext)

    i = 0
    for fi in fi_ext:
        H = getH(alpha, fi)
        eigs = H.eigenenergies()
        E0[i] = eigs[0]
        E1[i] = eigs[1]
        E2[i] = eigs[2]
        E3[i] = eigs[3]
        i += 1

    # Plot differences
    plt.plot(fi_ext, (E1 - E0) / 1e9)
    plt.plot(fi_ext, (E2 - E1) / 1e9)
    plt.plot(fi_ext, (E2 - E0) / 2e9)
    plt.legend(["$E_1 - E_0$", "$E_2 - E_1$", "$(E_2 - E_0)/2$"])
    
    # Plot eigenenergies
#     plt.plot(fi_ext, E0 / 1e9)
#     plt.plot(fi_ext, E1 / 1e9)
#     plt.plot(fi_ext, E2 / 1e9)
#     plt.plot(fi_ext, E3 / 1e9)

    plt.xlabel("$ \phi_{ext} = 2\pi \Phi_{ext}/\Phi_0$, radians")
    plt.ylabel("E, GHz")
    plt.title(r"$\alpha = $" + str(alpha))
    plt.xlim(4/5*const.pi, 6/5*const.pi)
```

```python
plt.figure(figsize=(8, 8))
# plt.subplot(121)
calc_and_plot_Eigenenergies(0.43)
# plt.subplot(122)
# calc_and_plot_Eigenenergies(0.6)
plt.tight_layout()
plt.savefig("eigenenergies_dif.png")
plt.show()
```
